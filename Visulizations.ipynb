```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('heart.csv')
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x="age",y="target")
plt.show()
```

```python
sns.barplot(data=df,x="sex",y="chol")
plt.show()
```

```python
sns.lineplot(data=df,x="age",y="target")
plt.show()
```

```python
sns.catplot(data=df,x="age",y="target")
plt.show()
```

```python
sns.countplot(x=df["target"])
```

```python
df.hist(figsize=(10,10))
```

```python
sns.pairplot(df, hue= 'target',vars = ['age', 'sex', 'cp', 'trestbps', 'chol' ])
```

```python
df.boxplot(figsize=(16,8))
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.corr(), annot = True, cmap='coolwarm', linewidths = 2)
```
